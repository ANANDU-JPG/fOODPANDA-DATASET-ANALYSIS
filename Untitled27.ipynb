```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
fd["customer_id"]=le.fit_transform(fd["customer_id"])
fd["gender"]=le.fit_transform(fd["gender"])
fd["age"]=le.fit_transform(fd["age"])
fd["delivery_status"]=le.fit_transform(fd["delivery_status"])
fd["city"]=le.fit_transform(fd["city"])
fd["signup_date"]=le.fit_transform(fd["signup_date"])
fd["order_id"]=le.fit_transform(fd["order_id"])
fd["order_date"]=le.fit_transform(fd["order_date"])
fd["category"]=le.fit_transform(fd["category"])
fd["restaurant_name"]=le.fit_transform(fd["restaurant_name"])
fd["dish_name"]=le.fit_transform(fd["dish_name"])
fd["rating_date"]=le.fit_transform(fd["rating_date"])
fd["payment_method"]=le.fit_transform(fd["payment_method"])
fd["churned"]=le.fit_transform(fd["churned"])
fd["last_order_date"]=le.fit_transform(fd["last_order_date"])
fd
```

```python
for col in dr.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=dr, x=col, hue='churned')
    plt.title(f'Count Plot of {col} by Churn Status')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
